### Load data

```python
# Helper functions

def xml_file(image_key, SAMPLE):
    """
    load in the xml file to see which files go in which spot
    create dataframe with file name and the corresponding row and col
    """
    filestring = os.path.join(SAMPLE, image_key + '.xml')
    xml_data = open(filestring, 'r').read()  # Read file
    data = xml_data.split('\n\t\t')
    data_sub = []
    for string in data:
        if 'ims' in string: 
            data_sub.append(string)

    df = pd.DataFrame(columns = ['COL','ROW','FILE','ABS_H','ABS_V'])
    for a in data_sub: 
        if 'COL' not in a: 
            continue
        b = a.split('COL="')[1]
        c = b.split('"')[0]
        col = int(c)

        b = a.split('ROW="')[1]
        c = b.split('"')[0]
        row = int(c)

        b = a.split('IMG_REGEX="')[1]
        c = b.split('"')[0]
        file = c

        b = a.split('ABS_H="')[1]
        c = b.split('"')[0]
        abs_h = int(c)

        b = a.split('ABS_V="')[1]
        c = b.split('"')[0]
        abs_v = int(c)

        df.loc[len(df.index)] = [col,row,file,abs_h,abs_v]
    
    return df
    
    
def load_ims(filename, channel='Channel 0'):
    ims = h5py.File(filename, 'r')
    x = ims['DataSet']['ResolutionLevel 0']['TimePoint 0'][channel]['Data'][:][0]
    ims.close()
    return x


def load_tiles(
    meta_df, 
    sample_id,
    channel='Channel 0'
):
    """
    Load image tiles from metadata dataframe
    """
    
    # Compute height & width for each IMS tile
    ncol = meta_df['COL'].max()
    nrow = meta_df['ROW'].max()
    w = ((meta_df['ABS_H'].max() - meta_df['ABS_H'].min()) / ncol).astype(np.int32)
    h = ((meta_df['ABS_V'].max() - meta_df['ABS_V'].min()) / nrow).astype(np.int32) 

    tiled = np.zeros((h*(nrow+1), w*(ncol+1)))  # tiled image
    
    for i in range(meta_df.shape[0]):
        # Coordinate order: right to left, bottom to top
        dY, dX = tiled.shape[0] - meta_df.iloc[i]['ABS_V'] - h, tiled.shape[1] - meta_df.iloc[i]['ABS_H'] - w
        fname = meta_df.iloc[i]['FILE']
        curr_tile = load_ims(os.path.join(sample_id, fname), channel=channel)[:h, :w]        
        tiled[dY:dY+h, dX:dX+w] = curr_tile

    return tiled


def to_numpy(filename):
    ims = h5py.File(filename, 'r')
    channel0 = ims['DataSet']['ResolutionLevel 0']['TimePoint 0']['Channel 0']['Data'][()]
    channel1 = ims['DataSet']['ResolutionLevel 0']['TimePoint 0']['Channel 1']['Data'][()]
    channel2 = ims['DataSet']['ResolutionLevel 0']['TimePoint 0']['Channel 2']['Data'][()]
    channel3 = ims['DataSet']['ResolutionLevel 0']['TimePoint 0']['Channel 3']['Data'][()]
    ims.close()
    return [channel0[0], channel1[0], channel2[0], channel3[0]]


def normalize(img, q0, q1):
    """Normalize image with quantile transformation"""
    assert 0 <= q0 < q1 <= 1, "Invalid rescale range"

    if img.ndim == 2:
        img_scaled = img.copy()
        vmin, vmax = np.quantile(img, [q0, q1])
        img_scaled[img_scaled < vmin] = vmin
        img_scaled[img_scaled > vmax] = vmax
        return (img_scaled - img_scaled.min()) / (img_scaled.max() - img_scaled.min())  # Convert to [0, 1]
    
    elif img.ndim == 3:
        img_scaled = img.copy()

        for i, chan in enumerate(img):
            vmin, vmax = np.quantile(chan, [q0, q1])
            img_scaled[i][chan < vmin] = vmin
            img_scaled[i][chan > vmax] = vmax

            img_scaled[i] = (img_scaled[i] - img_scaled[i].min()) / (img_scaled[i].max() - img_scaled[i].min())  # Convert to [0, 1]
        return img_scaled
    
    else:
        raise ValueError("Image should be either 2D (single channel) or 3D (multi channel)")


def trim_edges(img, width):
    """Inplace operation to trim edge pixels of the input img to be divisible by `width`"""
    nrows_to_rm, ncols_to_rm = img.shape[-2]%width, img.shape[-1]%width
    img = img[:, :-nrows_to_rm, :-ncols_to_rm]
    return img
```

### Compute Affine Transformation Matrix

```python
# Try aligning via SIFT & obtain the Global affine transformation M for all WF -> CF registration

def sift_registration(img_src, img_dst):
    img_src = img_src.astype(np.uint8)
    img_dst = img_dst.astype(np.uint8)
    
    sift = cv2.SIFT_create()
    pts_src, des_src = sift.detectAndCompute(img_src, None)
    pts_dst, des_dst = sift.detectAndCompute(img_dst, None)
    
    matcher = cv2.BFMatcher()
    matches = matcher.knnMatch(des_src, des_dst, k=2)
    
    good_matches = []
    for m, n in matches:
        if m.distance < 0.75*n.distance:
            good_matches.append(m)
    
    pts1 = np.float32([pts_src[m.queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
    pts2 = np.float32([pts_dst[m.trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)
    
    H, _ = cv2.findHomography(pts1, pts2, cv2.RANSAC, 5.0)
    img_src_warped = cv2.warpPerspective(img_src, H, (img_dst.shape[1], img_dst.shape[0]))

    return img_src_warped, H


WF_chan0 = tifffile.imread('../../DLI/tiled_channels/WF_chan_0.ome.tiff')
CF_chan0 = tifffile.imread('../../DLI/tiled_channels/CF_chan_0.ome.tiff')

_, M = sift_registration(WF_chan0, CF_chan0)

del WF_chan0, CF_chan0
print(M)
```

```python
# Global affine transformation matrix applied onto each cycle
M = np.array([[  1.00056508e+00,  -3.08802841e-04,   7.87117348e+02],
              [  1.07292347e-03,   9.99416632e-01,  -1.26591025e+02],
              [  3.56765465e-07,  -1.57529225e-07,   1.00000000e+00]])

M
```

### Full WF->CF affine transformation

```python
def multi_cycle_registration(
    guide_df, 
    M,
    data_path,
    sample_id,
    res_path,
    scale=0.25,
    tile_width=2048
):
    """Multi-cycle registration function"""
    sample_id = os.path.join(data_path, sample_id)
    wf_guides = guide_df['WF (750)'].to_list()
    cf_guides = guide_df['CF'].to_list()
    assert len(wf_guides) == len(cf_guides), "CF & WF should have the same cycle number"

    wf_chans = ['Channel 0', 'Channel 1']
    cf_chans = ['Channel 0', 'Channel 1', 'Channel 2']

    for (wf_key, cf_key) in zip(wf_guides, cf_guides):
        print('Aligning WF ({}) towards CF({})...'.format(wf_key, cf_key))
        print('================================')

        #--------------------------------------------------
        # (1). Load WF & CF whole-slide image from tiles
        #--------------------------------------------------
        wf_df = xml_file(image_key=wf_key, SAMPLE=sample_id)
        wf_img = np.array([
            load_tiles(wf_df, sample_id=sample_id, channel=chan)
            for chan in wf_chans
        ])
        print('\tWF raw image:', wf_img.shape)
        wf_img = rescale(wf_img, scale=scale, channel_axis=0, preserve_range=True)
        wf_img = normalize(wf_img, 0., 0.98)
        print('\tWF downscaled image:', wf_img.shape)
                       
        cf_df = xml_file(image_key=cf_key, SAMPLE=sample_id)
        cf_img = np.array([
            load_tiles(cf_df, sample_id=sample_id, channel=chan)
            for chan in cf_chans
        ])
        print('\tCF raw image:', cf_img.shape)
        cf_img = normalize(cf_img, 0., 0.98)
        H, W = np.round(cf_img.shape[-2]*scale).astype(np.int16), np.round(cf_img.shape[-1]*scale).astype(np.int16) # downscaled dimension for warped image

        #-----------------------------------------------------------
        # (2). Warp WF downsaled image w/ precompted affine matrix
        #-----------------------------------------------------------
        wf_img_warped = np.zeros((len(wf_chans), H, W), dtype=np.uint8)
        for i, chan in enumerate(wf_img):
            wf_img_warped[i] = cv2.warpPerspective(np.round(chan*255).astype(np.uint8), M, (W, H))
        wf_img_warped = wf_img_warped.astype(np.float64)
        wf_img_warped = (wf_img_warped-wf_img_warped.min()) / (wf_img_warped.max()-wf_img_warped.min())
        wf_img_warped = rescale(wf_img_warped, scale=1/scale, channel_axis=0, preserve_range=True)  # Rescale WF image back to original resolution

        assert cf_img.min() == wf_img_warped.min() == 0 and cf_img.max() == wf_img_warped.max() == 1, \
            "Image should be scaled to [0, 1]"

        #----------------------------------------------------------------------
        # (3). Save CF & warped WF image into tiles divisible by `tile_width`
        #----------------------------------------------------------------------
        print('\n\tSaving multiplexed CF & WF (warped) images...')
        wf_outdir = os.path.join(res_path, wf_key)
        if not os.path.isdir(wf_outdir):
            os.makedirs(wf_outdir, exist_ok=True)
            
        cf_outdir = os.path.join(res_path, cf_key)
        if not os.path.isdir(cf_outdir):
            os.makedirs(cf_outdir, exist_ok=True)
        
        wf_img_warped = trim_edges(wf_img_warped, tile_width)
        cf_img = trim_edges(cf_img, tile_width)
        print('\tWF trimmed image:', wf_img_warped.shape)
        print('\tCF trimmed image:', cf_img.shape)
        
        nrows, ncols = int(cf_img.shape[-2] / tile_width), int(cf_img.shape[-1] / tile_width)     
        for i in range(nrows):
            for j in range(ncols):
                idx0 = str(i) if i >= 10 else '0'+str(i)
                idx1 = str(j) if j >= 10 else '0'+str(j)
                if i % 5 == 0 and j % 5 == 0:
                    print('\t\tSaving CF & WF tile_{0}_{1}...'.format(idx0, idx1))

                # Save WF tile
                tifffile.imwrite(
                    os.path.join(wf_outdir, 'WF_tile_{0}_{1}.tif'.format(idx0, idx1)), 
                    wf_img_warped[:, tile_width*i:tile_width*(i+1), tile_width*j:tile_width*(j+1)]
                )   
        
                # Save CF tile
                tifffile.imwrite(
                    os.path.join(cf_outdir, 'CF_tile_{0}_{1}.tif'.format(idx0, idx1)), 
                    cf_img[:, tile_width*i:tile_width*(i+1), tile_width*j:tile_width*(j+1)]
                )

        print('================================\n')
        del cf_img, wf_img, wf_img_warped
        gc.collect()

    return None
```
